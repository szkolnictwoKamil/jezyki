```python
import pandas as pd

df1 = pd.read_csv("C:/repo/Plant_1_Generation_Data.csv", sep = ',', parse_dates=['DATE_TIME'], index_col = 0)

df2 = pd.read_csv("C:/repo/Plant_2_Generation_Data.csv", sep = ',', parse_dates=['DATE_TIME'], index_col = 0)

df = pd.concat([df1, df2], axis=0)
df.dropna()
```

```python
df_specific = df[df['SOURCE_KEY'] == '1BY6WEcLGh8j5v7']
df_specific
```

```python
df_week = df_specific.loc['2020-05-14':'2020-05-21']
df_week

df.loc[:, "AVG"] = df.groupby(df.index)['AC_POWER'].mean()
df_avg = df.loc['2020-05-15':'2020-05-21']
```

```python
df_week.AC_POWER.plot()

df_avg.AVG.plot(color = 'r', xlabel = 'Day', ylabel = 'AC_POWER', figsize=(15,10))
```

```python
df_below_avg = df[df["AC_POWER"] < 0.8*df["AVG"]]
df_below_avg
```
